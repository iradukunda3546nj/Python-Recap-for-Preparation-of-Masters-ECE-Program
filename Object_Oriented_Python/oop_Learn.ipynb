# Problem

Suppose we are developing software for ABC Digital Bank.

The bank has different types of accounts:

Savings ,
Current Account,
Fixed Deposit Account

Question:

Do these accounts have anything in common?

Yes.

Every account has

Account Number,
Owner Name,
Balance

Every account can

Deposit money,
Withdraw money,
Display account information

Instead of writing these three things repeatedly in every class, we'll create one parent class.

````python
#Parent Class
class BankAccount:

    def __init__(self, account_number, owner, balance=0):
        self.account_number = account_number
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            print("Deposit amount must be greater than zero.")
            return

        self.balance += amount
        print(f"${amount} deposited successfully.")

    def withdraw(self, amount):

        if amount <= 0:
            print("Invalid withdrawal amount.")
            return

        if amount > self.balance:
            print("Insufficient balance.")
            return

        self.balance -= amount
        print(f"${amount} withdrawn successfully.")

    def display_account(self):

        print("\n------ Account Information ------")
        print(f"Owner          : {self.owner}")
        print(f"Account Number : {self.account_number}")
        print(f"Balance        : ${self.balance:.2f}")
````

Notice something.

This class knows nothing about

Savings
Current
Fixed Deposit

It only contains everything common to every bank account.

Savings Account

A savings account has one extra feature.

It earns interest.

````python
class SavingsAccount(BankAccount):

    def add_interest(self, rate):

        interest = self.balance * rate / 100
        self.balance += interest

        print(f"Interest of ${interest:.2f} added.")
````

Look carefully.

We never rewrote

deposit()
withdraw()
display_account()

because SavingsAccount automatically inherited them.

Current Account

A current account allows an overdraft.

Let's keep it simple.

````python
class CurrentAccount(BankAccount):

    def __init__(self, account_number, owner, balance=0, overdraft_limit=500):

        super().__init__(account_number, owner, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):

        if amount <= 0:
            print("Invalid withdrawal.")
            return

        if amount > self.balance + self.overdraft_limit:
            print("Overdraft limit exceeded.")
            return

        self.balance -= amount
        print(f"${amount} withdrawn successfully.")
````

Notice something amazing.

This class inherits everything, but it changes only one method:

withdraw()

We'll later learn this is called overriding.

For now, just notice inheritance allowed us to reuse almost everything.

````python
#Creating Objects
savings = SavingsAccount(
    account_number="S001",
    owner="Jean Claude",
    balance=1000
)

current = CurrentAccount(
    account_number="C001",
    owner="Alice",
    balance=500
)
````

````python
savings.display_account()
current.deposit(1200)
current.withdraw(2000)
current.deposit(300)
````

### Python __init__() Method
--- 
The __init__() Method
All classes have a built-in method called __init__(), which is always executed when the class is being initiated.

The __init__() method is used to assign values to object properties, or to perform operations that are necessary when the object is being created.

#### Example
Create a class named Person, use the __init__() method to assign values for name and age:



````python
class Person:

    def __init__(self,name,age):
        self.name=name
        self.age=age
    def get_name(self):
        print(f"My name is {self.name} and I am {self.age} years old")

#creating an object
person1=Person("Clever",25) #constructing an object using constructor() method
person1.get_name() #object referencing to a method
print(person1.name) #an object referencing to one of its states
print(person1.age)
````

Note: The __init__() method is called automatically every time the class is being used to create a new object.

## Why Use __init__()?
Without the __init__() method, you would need to set properties manually for each object:

````python
class Person:
    pass


p1=Person()
p1.name="Clever"
p1.age=25
print(f"My name is {p1.name} and I am {p1.age} years old.")
````

### Class Properties vs Object Properties
Properties defined inside __init__() belong to each object (instance properties).

Properties defined outside methods belong to the class itself (class properties) and are shared by all objects:

````python
class Person:
  species = "Human" # Class property and it's shared by all objects. i.e. it's global

  def __init__(self, name):
    self.name = name # Instance/object property

p1 = Person("Emil")
p2 = Person("Tobias")

print(p1.name)
print(p2.name)
print(p1.species)
print(p2.species)
````

# Working on the Inheritance


````python
#Scenario: Let's say we want to have a program and it will include Students,Lectures,Admistrative peoples
#for inheritance we look for what those classes have in common and put in in the parent/base class Let's call it Person

class Person:
    def __init__(self,fName,lName,age,tel):
        self.fName=fName
        self.lName=lName
        self.age=age
        self.tel=tel
    
    def get_info(self):
        print("-------------Info---------")
        print("Name:",self.fName, self.lName)
        print("Aged:",self.age)
        print("Tel:",self.tel)


class Student(Person):
    def stud_register(self):
        print(f"student named {self.fName} {self.lName} registered successfully !")

class Lecturer(Person):
    pass

#Let's create an object for student
S1=Student("Jean Claude","Iradukunda",26,"+250781083227")
S1.get_info()
S1.stud_register()

#Let's create an object for the Lecturer   //Now you can see that all classes (Lecturer,Student) are inheriting get_info() method from parent class named Person
l1=Lecturer("Dr. Omar", "Gatera",56,"+250734880452")
l1.get_info()
````

## Adding __init__() to the child/derived class
---
as you might have seen in the above example,all child class inherit the __init__() from the parent class, once we re-define the __init__() in the child class , it will overide that one from the parent class as shown in the examples below.

````python
class Accountant(Person):
    def __init__(self,fName,lName,age,tel,city): #this __init__() will overrides the parent''s __init__()
        super().__init__(fName,lName,age,tel) #then this is a way to regain the __init()__ from parent class by using Classname dot function i.e. Person.__init..in this case or use super().__init__ ,it also inherit methods and properties from the parent class
        #Note: when you used super().__init__() to inherit from parent class, you no longer use self parameter inside i.e you omit it,  like  super().__init__(fName,lName,..)
        self.city=city #Look, we use super() to say that the child class i.e. Accountant will inherit from parent class but used when you want to add other parameters i.e."city" in this case

acc1=Accountant("Mwiza","Aliane",45,"+0781083227","Kigali")
print(acc1.city)
acc1.get_info()
````

## Python Encapsulation
---
Encapsulation is about protecting data inside a class.

It means keeping data (properties) and methods together in a class, while controlling how the data can be accessed from outside the class.

This prevents accidental changes to your data and hides the internal details of how your class works.

#### Private Properties
In Python, you can make properties private by using a double underscore __ prefix:

Example
Create a private class property named __age:

````python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age # Private property

p1 = Person("Emil", 25)
print(p1.name)
#print(p1.__age) # This will cause an error because the property age is private
````
